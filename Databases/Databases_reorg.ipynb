```python
import pandas as pd
from datetime import date, datetime, timedelta, time
import numpy as np
pd.options.mode.chained_assignment = None
```

### Events database

```python
events = pd.read_excel("Events_database.xlsx", index_col=0, na_values=["Не вказано", "Розпродано"]).astype({"Start time": str})
```

```python
months_dict = {"січ.": "01","лют.": "02", "бер.": "03", "квіт.": "04", "трав.": "05", "черв.":"06", "лип.":"07", "серп.":"08", "вер.":"09", "жовт.":"10", "лист.":"11", "груд.":"12"}
events["Start time"] = events.apply(lambda row: row["Date"][:2] + "." + months_dict[row["Date"][3:].strip()] + ".2024" + " " + row["Start time"], axis="columns").map(lambda x: datetime.strptime(x, "%d.%m.%Y %H:%M"))
events.drop("Date", inplace = True, axis="columns")
events.sort_values(by="Start time")
mean_dur = round(events["Duration"].dropna().mean())
events["Duration"] = events["Duration"].fillna(mean_dur).astype("int64")
events.head()
```

```python
events["Prices lower"] = events["Price range ₴"].map(lambda x: x[:x.find("-")].strip() if "-" in x else x)
events["Prices upper"] = events["Price range ₴"].map(lambda x: x[x.find("-") + 1:].strip() if "-" in x else x)
events.drop("Price range ₴", inplace=True,axis="columns")
events = events.astype({"Prices lower": "int32", "Prices upper": "int32"})
events
```

```python
events["End time"] = events.apply(lambda row: row["Start time"] + timedelta(minutes=row["Duration"]), axis="columns")
events
```

```python
events.to_csv("Events.csv")
```

```python
menu["Price"] = menu["Price"].map(lambda x: float(x.replace("\xa0","").replace(",",".").replace("₴","")))
menu
```

### Average price and quartiles

```python
rest_final["Average price"] = menu.groupby(["ID_Rest"]).apply(lambda df: round(df[(df["Price"] >= df["Price"].describe().iloc[4]) & (df["Price"] <= df["Price"].describe().iloc[6])]["Price"].mean()))
quart = rest_final["Average price"].describe()
rest_final["Overall price level"] = rest_final["Average price"].map(lambda level: "Low" if level < quart.iloc[4] else "Medium-Low" if level < quart.iloc[5] else "Medium-High" if level < quart.iloc[6] else "High")
rest_final
```

### Categories price

```python
categories_prices_mult = menu.groupby(["ID_Rest","Category"])["Price"].agg([min, max, "mean", "count"]).astype({"min":"int64", "max":"int64", "mean":"int64", "count":"int64"}).rename(columns={"min":"Min price", "max":"Max price", "mean": "Mean price", "count": "Number of items"})
categories_prices = categories_prices_mult.reset_index().set_index("ID_Rest")
categories_prices_mult
```
